# Paridade de funções

Este notebook servirá para a implementação de alguns exemplos de funções pares e ímpares, que mostrará a simetria de funções para o ensino de séries de Fourier dentro do curso de Introdução ao Processamento de Dados Sísmicos. 

## O que são funções pares e ímpares?

**Definição 1**: Uma função $f: \mathbb{R} \rightarrow \mathbb{R}$ é dita par se

$$ f(-x) = f(x).$$ 

**Definição 2**: Uma função $f: \mathbb{R} \rightarrow \mathbb{R}$ é dita ímpar se

$$ f(-x) = -f(x).$$

As paridade de funções nos fornece informações sobre a simetria de uma dada função. Uma função par será simétrica em relação ao eixo x, já uma função ímpar será simétrica em relação a origem do sistema de coordenadas. 

Abaixo serão demonstradas algumas propriedades que versam sobre a paridade das funções:

* A única função par e ímpar simultaneamente é a função nula
* Existem funções que não são pares nem impares
* A soma de duas funções de mesma paridade, mantém-se a mesma paridade
* O produto de funções de mesma paridade é uma função par
* O produto de funções de paridades diferentes é uma função ímpar
* A derivada de uma função par é uma função ímpar, e vice-versa

## Importando as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plotando algumas funções de uma variável

### Função linear

$$ f(x) = x$$

```python
x = np.linspace(-5, 5, 101)
y = 2*x
```

```python
plt.figure(figsize=(10,6))
plt.title('Linear function',fontsize=16)
plt.plot(x,y,'b-')
plt.axhline(y=0, color="black", linestyle="--")
plt.axvline(x=0, color="black", linestyle="--")
plt.xlim(np.min(x),np.max(x))
plt.grid(True)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.show()
```

### Função Quadrática

$$ f(x) = x^2 $$

```python
x = np.linspace(-5, 5, 101)
y = x**2
```

```python
plt.figure(figsize=(10,6))
plt.title('Quadratic function',fontsize=16)
plt.plot(x,y,'b-')
plt.axhline(y=0, color="black", linestyle="--")
plt.axvline(x=0, color="black", linestyle="--")
plt.xlim(np.min(x),np.max(x))
plt.grid(True)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.show()
```

### Função seno

$$ f(x) = \sin(x) $$

```python
x = np.linspace(-10,10,201)
y = np.sin(x)
```

```python
plt.figure(figsize=(10,6))
plt.title('Sine function',fontsize=16)
plt.plot(x,y,'b-')
plt.axhline(y=0, color="black", linestyle="--")
plt.axvline(x=0, color="black", linestyle="--")
plt.xlim(np.min(x),np.max(x))
plt.xticks(ticks=[i*np.pi for i in range(-2, 3)], 
           labels=[('%d $\pi$' % i) for i in range(-2, 3)])
plt.grid(True)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x (radians)',fontsize=16)
plt.show()
```
