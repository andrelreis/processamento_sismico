## As Séries de Fourier

**Definição**: Uma função $f: \mathbb{R} \rightarrow \mathbb{R}$ periódica de período $T \in \mathbb{R}$ pode ser expressa de acordo com a equação abaixo:


$$ f(x) = a_0 + \sum_{n=1}^{\infty} \left[a_n \cos(\lambda_n x) + b_n \sin(\lambda_n x)\right],$$

em que,

$$ a_0 = \frac{1}{2L} \int_{-L}^{L} f(x) dx, $$

$$ a_n = \frac{1}{L} \int_{-L}^{L} f(x) \cos(\lambda_n x) dx, $$

e

$$ b_n = \frac{1}{L} \int_{-L}^{L} f(x) \sin(\lambda_n x) dx, $$
nos quais $a_0$, $a_n$ e $b_n \in \mathbb{R}$ e 

$$ \lambda_n = \frac{n \pi}{L}.$$

Para que a Série de Fourier seja válida a função $f(x)$ deve ser periódica de período $T = 2L$. Além disso, $a_n$ e $b_n$ tem existência garantida se a função $f(x)$ for integrável ao longo do intervalo $[-L,L]$.

## Calculando as Séries de Fourier da funções Dente de Serra e Quadrada

### Função Dente de Serra

$$f(x) = \frac{1}{\pi} x, \, -\pi < x < \pi.$$

em que sua série de Fourier é representada por:

$$ f(x) = -\frac{2}{\pi} \sum_{n=1}^{\infty} \frac{(-1)^{n-1}}{n} \sin(2nx).$$

```python
x = np.linspace(-10, 10, 301)
y = sawtooth(t=x-np.pi,width=1)
```

```python
n = 30
bn_saw = func.sawtooth_bn(n)
fx = func.fourier_series(x,a0=0,an=np.zeros_like(bn_saw),bn=bn_saw)
```

```python
plt.figure(figsize=(10,6))
plt.title('Sawtooth function',fontsize=16)
plt.plot(x,y,'b-')
plt.plot(x,fx,'r-')
plt.axhline(y=0, color="black", linestyle="--")
plt.xlim(np.min(x),np.max(x))
plt.xticks(ticks=[i*np.pi for i in range(-3, 4)], 
           labels=[('%d $\pi$' % i) for i in range(-3, 4)])
plt.grid(True)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x (radians)',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.show()
```

### Função Quadrada

$$f(x) =\begin{cases}
 & 1 \, \, \, \,\quad \text{se} \quad 0 \le x \le \pi \\
 & -1\quad \text{se} \quad  -\pi \le x < 0  
\end{cases},$$

em que a Série de Fourier associada a função quadrada será igual a:

$$ f(x) = \frac{4}{\pi} \sum_{n \, odd}^{\infty} \frac{1}{n} \sin(nx).$$

```python
x = np.linspace(-10,10,301)
y = 0.5*(square(t=x,duty=0.5))
```

```python
n = 40
bn_sqr = func.square_bn(n)
fx = func.fourier_series(x,a0=0,an=np.zeros_like(bn_sqr),bn=bn_sqr)
```

```python
plt.figure(figsize=(10,6))
plt.title('Square Function',fontsize=16)
plt.plot(x,y,'b-')
plt.plot(x,fx,'r-')
plt.axhline(y=0, color="black", linestyle="--")
plt.xlim(np.min(x),np.max(x))
plt.xticks(ticks=[i*np.pi for i in range(-3, 4)], 
           labels=[('%d $\pi$' % i) for i in range(-3, 4)])
plt.grid(True)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x (radians)',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.show()
```
