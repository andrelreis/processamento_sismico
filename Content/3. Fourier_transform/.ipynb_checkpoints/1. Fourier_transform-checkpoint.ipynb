# Transformada de Fourier

Este notebook servirá para a definição do que é uma transformada de Fourier e demonstrar algumas propriedades da transformada. Este notebook servirá para a continuação do estudo de Séries de Fourier fornecido no curso de Introdução ao Processamento de Dados Sísmicos.   

## O que é uma Transformada de Fourier?

**Definição 1**: Uma função $f: \mathbb{R} \rightarrow \mathbb{C}$ é 

periódica de período $T \in \mathbb{R}$ se para todo $x \in \mathbb{R}$ termos que,

$$ f(x + T) = f(x).$$ 

Toda função é uma função periódica de período zero. Contudo, considerarei aqui funções com período $T \neq 0$. Funções constantes são funções periódicas de qualquer período e, portanto, consideraremos funções que não são constantes. Se $T$ é período de $f$, então $nT$ também será período de $f$ para todo $n \in \mathbb{N}$. 

**Definição 2**: O menor período $T \neq 0$, se existir, tal que $f(x + T) = f(x)$, para todo $x \in \mathbb{R}$, é dito período fundamental de $f$.

## Importando as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.signal import sawtooth, square
```

## Plotando as funções periódicas de período $2\pi$

### Função Sawtooth

```python
x = np.linspace(-10, 10, 101)
y = sawtooth(t=x-np.pi,width=1)
```

```python
plt.figure(figsize=(10,6))
plt.title('Sawtooth function',fontsize=16)
plt.plot(x,y,'b-')
plt.axhline(y=0, color="black", linestyle="--")
plt.xlim(np.min(x),np.max(x))
plt.xticks(ticks=[i*np.pi for i in range(-3, 4)], 
           labels=[('%d $\pi$' % i) for i in range(-3, 4)])
plt.grid(True)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x (radians)',fontsize=16)
plt.show()
```

### Função Odd Square

```python
x = np.linspace(-10,10,101)
y = square(t=x,duty=0.5)
```

```python
plt.figure(figsize=(10,6))
plt.title('Square Function',fontsize=16)
plt.plot(x,y,'b-')
plt.axhline(y=0, color="black", linestyle="--")
plt.xlim(np.min(x),np.max(x))
plt.xticks(ticks=[i*np.pi for i in range(-3, 4)], 
           labels=[('%d $\pi$' % i) for i in range(-3, 4)])
plt.grid(True)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x (radians)',fontsize=16)
plt.show()
```

## Plotando uma função periódica de período $4 \pi$

```python
x = np.linspace(-20,20,201)
y = np.cos(x) + np.cos(x/2)
```

```python
plt.figure(figsize=(10,6))
plt.title('Cosine stacking',fontsize=16)
plt.plot(x,y,'b-')
plt.axhline(y=0, color="black", linestyle="--")
plt.xlim(np.min(x),np.max(x))
plt.xticks(ticks=[i*np.pi for i in range(-6, 6)], 
           labels=[('%d $\pi$' % i) for i in range(-6, 6)])
plt.grid(True)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x (radians)',fontsize=16)
plt.show()
```
